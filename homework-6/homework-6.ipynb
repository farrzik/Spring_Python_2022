6.  Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.

```python
def func_i_plus_j(n):
    a = np.zeros((n,n))
    a[0,:] = np.arange(1, n+1)
    for i in range(1, n):
        a[i,:] = a[i-1,:]+1
    return a
func_i_plus_j(5)
```

10. Найдите сумму всех элементов, сумму внутри строк, сумму внутри столбцов, а также среднее значение, дисперсию и стандартное отклонение чисел для каждой строки матрицы $A$. (Подобно тому, как sorted имеет необязательный аргумент key=, многие функции Numpy имеют необязательный аргумент axis=)


```python
a = np.random.random(size = (5,5))
a_sum = a.sum()
summa_strok = np.zeros(5)
summa_strok[:] = a[:].sum(axis = 1)
summa_stolbcov = np.zeros(5)
summa_stolbcov[:] = a[:].sum(axis = 0)
mean_str = np.zeros(5)
mean_str[:] = a[:].mean(axis = 1)
var_str = np.zeros(5)
var_str[:] = a[:].var(axis = 1)
std_str = np.zeros(5)
std_str[:] = a[:].std(axis = 1)
print('Сумма элементов: ', a_sum, 'Сумма строк: ',summa_strok, 'Сумма столбцов: ', summa_stolbcov, 
      'Среднее по строкам: ', mean_str, 'дисперсия по строкам: ', var_str, 'Станд. Отклонение по строкам: ', std_str, sep = '\n')
```

15.	Посчитайте детерминант и ранг матриц $A$, $B$ и $C$, посмотрите на спектр (набор собственных значений) матриц $B$ и $C$, какие выводы вы можете сделать?  
(подсказка: используйте готовые функции из библиотеки np.linalg). 


```python
a_det = np.linalg.det(a)
b_det = np.linalg.det(b)
c_det = np.linalg.det(c)
b_sz = np.linalg.eig(b)
c_sz = np.linalg.eig(c)
print('Детерминант А:', a_det,'Детерминант В:',b_det,'Детерминант С:',c_det,'Соб. значения В',b_sz,
      'Соб. значения С',c_sz, sep = '\n')
print('Вывод - набор собственных значений у матриц В и С по значениям аналогичен, но по порядку различен')
```
